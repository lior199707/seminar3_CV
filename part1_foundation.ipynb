## H letter example

```python
# Create the H letter as a NumPy array (8x8 grayscale image)
h_letter = np.array([
    [255, 0, 255, 255, 255, 255, 0, 255],
    [255, 0, 255, 255, 255, 255, 0, 255],
    [255, 0, 255, 255, 255, 255, 0, 255],
    [255, 0,  0,   0,   0,   0,  0, 255],
    [255, 0,  0,   0,   0,   0,  0, 255],
    [255, 0, 255, 255, 255, 255, 0, 255],
    [255, 0, 255, 255, 255, 255, 0, 255],
    [255, 0, 255, 255, 255, 255, 0, 255]
], dtype=np.uint8)

# Display information about the array
print("Array shape:", h_letter.shape)
print("Data type:", h_letter.dtype)
print("\nArray content:")
print(h_letter)

# Display the image with OpenCV (works in standard Python scripts)
# cv2.imshow('H Letter', h_letter_scaled)
# print("\nPress any key to close the window...")
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display using Matplotlib (works in Jupyter Notebook)
plt.figure(figsize=(6, 6))
plt.imshow(h_letter, cmap='gray', vmin=0, vmax=255)
plt.title('H Letter')
plt.axis('off')  # Hide axis
plt.show()
```

## Access and Manipulate Image Pixels

```python
# image path
image_path = "images/marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# Access pixel at (0,0) location -> image[y, x]
(b, g, r) = image[0, 0]
print("Blue, Green and Red values at (0,0): ", format((b, g, r)))

# Manipulate pixels and show modified image
image[0:100, 0:200] = (255, 255, 0)

# # Display the image with OpenCV (works in standard Python scripts)
# cv2.imshow("Modified Image", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display using Matplotlib (works in Jupyter Notebook)
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Modified Mars Rover")
plt.axis('off')  # Hide axis
plt.show()
```

## Drawing Shapes

### Drawing Lines

```python
# image path
image_path = "images/marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# set start and end coordinates
start = (0, 0)
end = (image.shape[1], image.shape[0]) # (width, height) or (x, y)
# set the color in BGR
color = (255,0,0)
# set thickness in pixel
thickness = 4
cv2.line(image, start, end, color, thickness)

# display the modified image using OpenCV (works in standard Python scripts)
# cv2.imshow("Modified Image", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display using Matplotlib (works in Jupyter Notebook)
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Modified Mars Rover")
plt.axis('off')  # Hide axis
plt.show()
```

### Drawing Rectangles 

```python
# image path
image_path = "images/marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)
# set the start and end coordinates
# of the top-left and bottom-right corners of the rectangle
start = (63,48) # (x,y)
end = (208,232) # (x,y)
# Set the color and thickness of the outline
color = (0,255,0) # Green in BGR
thickness = 5
# Draw the rectangle
cv2.rectangle(image, start, end, color, thickness)

# Save the modified image with the rectangle drawn to it.
cv2.imwrite("rectangle.jpg", image)
# display the modified image using OpenCV (works in standard Python scripts)
# cv2.imshow("Rectangle", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display using Matplotlib (works in Jupyter Notebook)
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Modified Mars Rover")
plt.axis('off')  # Hide axis
plt.show()
```

```python
# create a new canvas
canvas = np.zeros((200, 200, 3), dtype = "uint8")
start = (10,10) # Top-left corner in form (x,y)
end = (100,100) # Bottom-right corner in form (x,y)
color = (0,0,255) # Red in BGR
thickness = -1 # Negative thickness means fill the rectangle
# Draw the rectangle
cv2.rectangle(canvas, start, end, color, thickness)

# Display the rectangle using OpenCV (works in standard Python scripts)
# cv2.imshow("Rectangle", canvas)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display using Matplotlib (works in Jupyter Notebook)
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title("Rectangle")
plt.axis('off')  # Hide axis
plt.show()
```

### Drawing Circles

```python
# create a new canvas
canvas = np.zeros((200, 200, 3), dtype = "uint8")
center = (100,100) # Center of the circle in form (x,y)
radius = 50
color = (0,0,255) # Red in BGR
thickness = 5
cv2.circle(canvas, center, radius, color, thickness)

# Display the circle using OpenCV (works in standard Python scripts)
# cv2.imshow("My Circle", canvas)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display using Matplotlib (works in Jupyter Notebook)
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title("My Circle")
plt.axis('off')  # Hide axis
plt.show()
```
